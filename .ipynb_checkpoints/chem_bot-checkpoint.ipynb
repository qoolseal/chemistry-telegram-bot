```python
# распаковываем файл с вопросами и ответами
quest = []
ans = []
with open('<path>/tests.txt', encoding="utf-8") as inp:
  for line in inp:
    s = line.split(';')
    quest.append(s[0])
    ans.append(s[1].strip())
#quest.append('')
#quest.append('Напиши "Ответы:", и через пробел пиши ответы, разделяя запятой с пробелом (Например, Ответы: 1, 2, 3')
stic = []
with open('<path>/stickers.txt', encoding="utf-8") as i:
  for line in i:
    stic.append(line.strip())

from telegram import InlineKeyboardButton, InlineKeyboardMarkup, Update
from telegram.ext import Updater, CallbackQueryHandler, ConversationHandler
updater = Updater('<token>', use_context=True)
      # telegram.ext.Updater получает все обновления/сообщения в автоматическом режиме.
dispatcher = updater.dispatcher
      # в объекте Dispatcher можно зарегистрировать обработчики разных типов,
      # которые будут сортировать полученные объектом Updater сообщения
from telegram.ext import MessageHandler, Filters  # модули необходимые для обработки сообщений
import logging  # Чтобы знать, когда и почему что-то не работает должным образом,
                # настроим модуль ведения журнала логов
logging.basicConfig(format='%(levelname)s - %(message)s', level =logging.DEBUG)
logger = logging.getLogger(__name__)

from telegram.ext import CommandHandler# импортируем обработчик CommandHandler, который фильтрует сообщения с командами

from chempy import chemistry
from chempy import balance_stoichiometry

import random

import pubchempy as pcp
from pubchempy import get_compounds

import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

import googletrans
from googletrans import Translator
translator = Translator()

import requests
from bs4 import BeautifulSoup

# Этапы/состояния разговора
FIRST, SECOND = range(2)
# Данные обратного вызова
ONE, TWO, THREE, FOUR, FIVE, SIX = range(6)

def start(update, context):
    """Вызывается по команде `/start`."""
    # Создаем `InlineKeyboard`, где каждая кнопка имеет
    # отображаемый текст и строку `callback_data`
    # Клавиатура - это список строк кнопок, где каждая строка,
    # в свою очередь, является списком `[[...]]`
    keyboard = [
        [
            InlineKeyboardButton("картинку", callback_data=str(FOUR)),
            InlineKeyboardButton("мини-тест", callback_data=str(FIVE)),
        ]
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)
    # Отправляем сообщение с текстом и добавленной клавиатурой `reply_markup`
    context.bot.send_message(chat_id=update.effective_chat.id,
        text='Привет, этот бот может:\n  Решить химическое уравнение, напиши: /pred и реагенты через пробел\nНапример, /pred H2SO4 NH3 \n  Уравнять химическое уравнение. Для этого напиши:\n/eq [хим уравнение в виде Al + O2 => Al2O3]\n  Нарисовать молекулу по базе данных. Для этого просто отправь название (или формулу) вещества на любом языке\nА ещё может отправить тебе:', reply_markup=reply_markup
    )
    # Сообщаем `ConversationHandler`, что сейчас состояние `FIRST`
    return FIRST

#def start(update, context):   # `bot.send_message` это метод Telegram API
    '''функция обрабатывает команду /start, в ответ на неё отправляет меню'''
    #context.bot.send_message(chat_id=update.effective_chat.id, # `update.effective_chat.id` - определяем `id` чата, откуда прилетело сообщение
      #text='Привет, этот бот может:\n    /eq [хим уравнение в виде Al + O2 => Al2O3] - Уравнять химическое уравнение\n    /draw [название молекулы] - Нарисовать молекулу по базе данных\n    /test - Выдать тебе список вопросов или\n    /pic - Отправить милую картиночку')

def equalize(update, context):
  '''функция для обработки команды с аргументами /equalize для уравнивания хим уравнений'''
  if context.args:
    text_eq = context.args
    equa = ' '.join(text_eq)
    text_eq = equa.split(' => ')
    pr = set(text_eq[1].split(' + '))
    re = set(text_eq[0].split(' + '))
    reac, prod = balance_stoichiometry(re, pr)
    re = []
    pr = []
    for k,v in reac.items():
      if v == 1:
        v = ''
      re.append(f'{v}{k}')
    for k,v in prod.items():
      if v == 1:
        v = ''
      pr.append(f'{v}{k}')
    re = ' + '.join(re)
    pr = ' + '.join(pr)
    text_eq = f'{re} => {pr}'

    context.bot.send_message(chat_id=update.effective_chat.id,
                                text=text_eq)
  else:
    context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Нет аргументов')
    context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Отправьте: /eq своё уравнение без коэффициентов')

def build_menu(buttons, n_cols,
               header_buttons=None,
               footer_buttons=None):
  '''шаблон для меню из кнопок'''

  menu = [buttons[i:i + n_cols] for i in range(0, len(buttons), n_cols)]
  if header_buttons:
      menu.insert(0, [header_buttons])
  if footer_buttons:
      menu.append([footer_buttons])
  return menu

def button(update, context):
  query = update.callback_query
  var = query.data
  list_cid = var.split(',')
  text = []
  n = len(list_cid)

  for i in range (n):
    text.append(list_cid[i])
    text.append('    ')
    c = pcp.Compound.from_cid(list_cid[i])
    syn = c.synonyms
    if len(syn) > 7:
      name = []
      for i in range (1, 7):
        n = syn[i].lower()
        if all(x.isalpha() or x.isspace() for x in n):
          g = translator.translate(n, dest='ru')
          g = f' ({g.text})'
          n += g
        name.append(n)
      name = '; '.join(name)
    else:
      name = []
      for n in syn:
        n = n.lower()
        if all(x.isalpha() or x.isspace() for x in n):
          g = translator.translate(n, dest='ru')
          g = f' ({g.text})'
          n += g
        name.append(n)
      name = '; '.join(name)
    text.append(name)
    text.append('\n')
    text.append('\n')
  context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Номера(cid) соединений и их названия:')
  context.bot.send_message(chat_id=update.effective_chat.id,
                                text=''.join(text))
def pred(update, context):
  '''функция, предсказывающая продукты реакции(берётся с сайта https://chemequations.com/ru/)'''
  if context.args:
    pr = context.args
    new = str(pr[0])
    for i in range(1,len(pr)):
      new+= '+%2B+' + str(pr[i])
    u = f'https://chemequations.com/ru/?s={new}&ref=input'
    response = requests.get(u)
    soup = BeautifulSoup(response.text, 'lxml')
    quotes = soup.find_all('span', class_ = 'group')
    soup1 = soup.find_all('a')
    href = soup1[11].get('href')
    try:
      if soup.find('caption').text == 'Существует несколько решений:':
        context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Существует несколько решений. Вот самое популярное:')
        u = 'https://chemequations.com/ru/' + str(href)
        response = requests.get(u)
        soup = BeautifulSoup(response.text, 'lxml')
        quotes = soup.find_all('span', class_ = 'group')
    except AttributeError:
      quotes = soup.find_all('span', class_ = 'group')
    ur = []
    for q in quotes:
      a = q.text
      if a.startswith('(') and a.endswith(')'):
        continue
      elif a in ur:
        continue
      else:
        ur.append(a)
    if len(ur) == len(pr):
      context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Не нашлось ответа. Попробуйте другую реакцию')
    else:
      ans = str(ur[0])
      ur = ur[1:]
      for i in range(len(ur)):
        if ur[i] in pr:
          ans += ' + ' + str(ur[i])
        #context.bot.send_message(chat_id=update.effective_chat.id,
                                #text=f'{ans}\n{pr}')
        if ur[i] not in pr:
          ans += ' => ' + str(ur[i])
          n = ur.index(ur[i])
          break

      if len(ur) != n:
        ur = ur[int(n + 1):]
        for i in range(len(ur)):
          ans += ' + ' + str(ur[i])

      context.bot.send_message(chat_id=update.effective_chat.id, text=ans)
      context.bot.send_message(chat_id=update.effective_chat.id, text='А теперь можете уравнять сами или при помощи команды /eq')
  else:
    context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Нет аргументов')
    context.bot.send_message(chat_id=update.effective_chat.id,
                                text='Отправьте: /pred и реагенты через пробел\nНапример, /pred H2SO4 NH3')



    
def test(update, context):
  '''функция обрабатывает команду /test, отправляя список вопросов'''

  text = '\n'.join(quest)
  keyboard = [
        [
            InlineKeyboardButton("4 3 2 1", callback_data=str(ONE)),
            InlineKeyboardButton("1 2 3 4", callback_data=str(TWO)),
            InlineKeyboardButton("4 2 3 4", callback_data=str(TWO)),
            InlineKeyboardButton("<= в меню", callback_data=str(SIX)),
        ]
    ]
  reply_markup = InlineKeyboardMarkup(keyboard)
  context.bot.send_message(chat_id=update.effective_chat.id, text = text, reply_markup=reply_markup)
  return FIRST

def pic(update, context):
  '''функция обработки команды для отправки рандомной картинки'''

  n = random.randint(1, 50)
  keyboard = [
        [
            InlineKeyboardButton("ещё картинку", callback_data=str(FOUR)),

        ]
    ]
  reply_markup = InlineKeyboardMarkup(keyboard)
  context.bot.send_photo(chat_id=update.effective_chat.id, photo=open(f'/home/qoolseal/{n}.jpg', 'rb'), reply_markup=reply_markup)
  return FIRST

def draw(update, context):
  '''функция обработки названий химических соединений'''
  mes = update.message.text
  name = ''.join(mes)
  name = str(name)
  translation = translator.translate(name, dest='en')
  name = translation.text
  list_comp = []
  list_cid = []
  for compound in get_compounds(name, 'name'):
    list_cid.append(compound.cid)
    list_comp.append(compound.isomeric_smiles)
  try:
    cid = list_cid[0]
  except IndexError:
    context.bot.send_message(chat_id=update.effective_chat.id,
          text='К сожалению, не нашли ничего подходящего. Попробуйте записать название по-другому или искать другое соединение.')
    return
  c = pcp.Compound.from_cid(cid)
  form = c.molecular_formula
  weight = c.molecular_weight
  iupac = c.iupac_name
  rus_iu = translator.translate(iupac, dest='ru')
  rus_iu = rus_iu.text
  context.bot.send_message(chat_id=update.effective_chat.id,
     text=f'Информация о молекуле:\nМолекулярная формула: {form}\nМолекулярная масса: {weight}\nНазвание по системе ИЮПАК: {iupac}\nИли: {rus_iu}')
  smiles = list_comp[0]
  mol = Chem.MolFromSmiles(smiles)
  Draw.MolToImageFile(mol,'imgPath.jpg',size=(500, 500), kekulize=True, wedgeBonds=True) # сохраняет тип Mol в файл
  context.bot.send_photo(chat_id=update.effective_chat.id, photo=open('imgPath.jpg', 'rb'))
  list_cid = ','.join(str(x) for x in list_cid)
  button_list = [InlineKeyboardButton("да", callback_data=list_cid)]# сборка клавиатуры из кнопок `InlineKeyboardButton`
  reply_mark = InlineKeyboardMarkup(build_menu(button_list, n_cols=1))
  context.bot.send_message(chat_id=update.effective_chat.id, text='Хотите увидеть остальные найденные варианты и синонимы названия?', reply_markup=reply_mark)# отправка клавиатуры в чат

#InlineKeyboardButton("<= в меню", callback_data=str(SIX)),
def unknown(update, context):
  '''функция обработки не распознных команд'''
  context.bot.send_message(chat_id=update.effective_chat.id,
                             text="Эту команду я не понимаю ):\n Попробуй отправить /start")
def ok(update, context):
  context.bot.send_message(chat_id=update.effective_chat.id,
                             text='Всё верно!')
  context.bot.send_sticker(chat_id=update.effective_chat.id,sticker=random.choice(stic))
def no(update, context):
  context.bot.send_message(chat_id=update.effective_chat.id,
                             text='Ой, где-то есть ошибка')
def sticker(update, context):

  context.bot.send_message(chat_id=update.effective_chat.id,
                             text='Крутой стикер!')
  context.bot.send_sticker(chat_id=update.effective_chat.id,sticker=random.choice(stic))
conv_handler = ConversationHandler(
        entry_points=[CommandHandler('start', start)],
        states={ # словарь состояний разговора, возвращаемых callback функциями
            FIRST: [
                CallbackQueryHandler(ok, pattern='^' + str(ONE) + '$'),
                CallbackQueryHandler(no, pattern='^' + str(TWO) + '$'),
                #CallbackQueryHandler(button, pattern='^' + str(THREE) + '$'),
                CallbackQueryHandler(pic, pattern='^' + str(FOUR) + '$'),
                CallbackQueryHandler(test, pattern='^' + str(FIVE) + '$'),
                CallbackQueryHandler(start, pattern='^' + str(SIX) + '$'),

            ],
            SECOND: [

            ],
        },
        fallbacks=[CommandHandler('start', start)],
    )

    # Добавляем `ConversationHandler` в диспетчер, который
    # будет использоваться для обработки обновлений
dispatcher.add_handler(conv_handler)

    # Добавляем `ConversationHandler` в диспетчер, который
    # будет использоваться для обработки обновлений
dispatcher.add_handler(conv_handler)
        #обработчик команды с аргументами для уравнивания реакции
eq_handler = CommandHandler('eq', equalize)
dispatcher.add_handler(eq_handler)
      # обработчик нажатия кнопки
updater.dispatcher.add_handler(CallbackQueryHandler(button))
      #обработчик команды с аргументами для рисования молекулы
draw_handler = MessageHandler(Filters.text, draw)
dispatcher.add_handler(draw_handler)
sticker_handler = MessageHandler(Filters.sticker, sticker)
dispatcher.add_handler(sticker_handler)
      # обработчик не распознных команд
unknown_handler = MessageHandler(Filters.command, unknown)
dispatcher.add_handler(unknown_handler)
updater.dispatcher.add_handler(CommandHandler('pred', pred))
updater.start_polling()# говорим объекту updater слушать сервера telegram
updater.idle()# обработчик нажатия Ctrl+C для остановки
updater.stop()
```
